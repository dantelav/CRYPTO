```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

df_eth['date'] = pd.to_datetime(df_eth['date'])
df_eth.sort_values('date', inplace=True)
data = df_eth[['open', 'high', 'low', 'close']].values

scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)

train_size = int(len(scaled_data) * 0.8)
train_data = scaled_data[:train_size]
test_data = scaled_data[train_size:]

def create_sequences(data, seq_length):
    X = []
    y = []
    for i in range(len(data)-seq_length):
        X.append(data[i:i+seq_length])
        y.append(data[i+seq_length])
    return np.array(X), np.array(y)

seq_length = 30
X_train, y_train = create_sequences(train_data, seq_length)
X_test, y_test = create_sequences(test_data, seq_length)

model = Sequential()
model.add(LSTM(64, activation='relu', input_shape=(seq_length, 4)))
model.add(Dense(4))
model.compile(optimizer='adam', loss='mse')
model.fit(X_train, y_train, epochs=10, batch_size=32)

train_loss = model.evaluate(X_train, y_train, verbose=0)
test_loss = model.evaluate(X_test, y_test, verbose=0)

train_predictions = model.predict(X_train)
test_predictions = model.predict(X_test)

train_predictions = scaler.inverse_transform(train_predictions)
test_predictions = scaler.inverse_transform(test_predictions)

last_sequence = np.expand_dims(test_data[-seq_length:], axis=0)
next_day_prediction = model.predict(last_sequence)
next_day_prediction = scaler.inverse_transform(next_day_prediction)[0]

fecha_prediccion = df_eth['date'].values[-1] + pd.DateOffset(days=1)
abre_prediccion = next_day_prediction[0]
max_prediccion = next_day_prediction[1]
min_prediccion = next_day_prediction[2]
cierra_prediccion = next_day_prediction[3]

percentage_change = (cierra_prediccion - abre_prediccion) / abre_prediccion * 100

# Calcular R2 para los conjuntos de entrenamiento y prueba
train_r2 = r2_score(y_train, train_predictions)
test_r2 = r2_score(y_test, test_predictions)

print("Fecha del dÃ­a predicho:", fecha_prediccion.date())
print("Apertura:", abre_prediccion)
print("MÃ¡ximo:", max_prediccion)
print("MÃ­nimo:", min_prediccion)
print("Cierre:", cierra_prediccion)
print("Porcentaje de cambio del precio a lo largo del dÃ­a:", percentage_change)
print("Valor con mayor precisiÃ³n:", ['MÃ¡ximo', 'MÃ­nimo', 'Cierre'][np.argmax(np.abs(next_day_prediction - test_data[-1]))])
print("EvaluaciÃ³n del modelo LSTM:")
print("Train Loss:", train_loss)
print("Test Loss:", test_loss)
print("Train R2 Score:", train_r2)
print("Test R2 Score:", test_r2)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Usar el DataFrame df_btc existente

# Convertir la columna 'date' a formato de fecha
df_btc['date'] = pd.to_datetime(df_btc['date'])

# Establecer la columna 'date' como Ã­ndice del DataFrame
df_btc.set_index('date', inplace=True)

# Crear una serie temporal para el valor de cierre
time_series = df_btc['close'].values.reshape(-1, 1)

# Normalizar los datos entre 0 y 1
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(time_series)

# Dividir los datos en conjunto de entrenamiento y prueba
train_data = scaled_data[:int(0.8 * len(scaled_data))]
test_data = scaled_data[int(0.8 * len(scaled_data)):]

# Crear secuencias de entrada y salida para el modelo LSTM
def create_sequences(data, sequence_length):
    X, y = [], []
    for i in range(len(data) - sequence_length):
        X.append(data[i:i+sequence_length])
        y.append(data[i+sequence_length])
    return np.array(X), np.array(y)

sequence_length = 7  # Longitud de la secuencia de entrada
X_train, y_train = create_sequences(train_data, sequence_length)
X_test, y_test = create_sequences(test_data, sequence_length)

# Construir el modelo LSTM
model = Sequential()
model.add(LSTM(units=64, input_shape=(sequence_length, 1)))
model.add(Dense(units=1))
model.compile(optimizer='adam', loss='mean_squared_error')

# Entrenar el modelo LSTM
model.fit(X_train, y_train, epochs=50, batch_size=32)

# Realizar predicciones en el conjunto de prueba
predictions = model.predict(X_test)
predictions = scaler.inverse_transform(predictions)

# Invertir la normalizaciÃ³n en los datos de prueba
y_test = scaler.inverse_transform(y_test)

# Calcular el coeficiente de determinaciÃ³n (RÂ²)
r2 = r2_score(y_test, predictions)
print("Coeficiente de determinaciÃ³n (RÂ²):", r2)

# Calcular el error cuadrÃ¡tico medio (MSE)
mse = mean_squared_error(y_test, predictions)
print("Error cuadrÃ¡tico medio (MSE):", mse)

# Calcular el error absoluto medio (MAE)
mae = mean_absolute_error(y_test, predictions)
print("Error absoluto medio (MAE):", mae)

# Graficar las predicciones y los valores reales
plt.plot(df_btc.index[int(0.8 * len(df_btc)) + sequence_length:], y_test, label='Valores reales')
plt.plot(df_btc.index[int(0.8 * len(df_btc)) + sequence_length:], predictions, label='Predicciones')
plt.xlabel('Fecha')
plt.ylabel('Valor de cierre')
plt.legend()
plt.show()
```
