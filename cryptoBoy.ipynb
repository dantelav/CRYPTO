```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Convertir la columna 'date' en tipo datetime
df_eth['date'] = pd.to_datetime(df_eth['date'])
df_eth.sort_values('date', inplace=True)

# Obtener los valores de las características
data = df_eth[['open', 'high', 'low', 'close']].values

# Escalar los datos
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)

# Dividir los datos en entrenamiento y prueba
train_size = int(len(scaled_data) * 0.8)
train_data = scaled_data[:train_size]
test_data = scaled_data[train_size:]

# Función para crear secuencias de entrenamiento
def create_sequences(data, seq_length):
    X = []
    y = []
    for i in range(len(data)-seq_length):
        X.append(data[i:i+seq_length])
        y.append(data[i+seq_length])
    return np.array(X), np.array(y)

seq_length = 30
X_train, y_train = create_sequences(train_data, seq_length)
X_test, y_test = create_sequences(test_data, seq_length)

# Definir el modelo LSTM
model = Sequential()
model.add(LSTM(64, activation='relu', input_shape=(seq_length, 4)))
model.add(Dense(4))
model.compile(optimizer='adam', loss='mse')

# Entrenar el modelo
model.fit(X_train, y_train, epochs=10, batch_size=32)

# Evaluar el modelo
train_loss = model.evaluate(X_train, y_train, verbose=0)
test_loss = model.evaluate(X_test, y_test, verbose=0)

# Hacer predicciones
train_predictions = model.predict(X_train)
test_predictions = model.predict(X_test)

# Invertir las transformaciones de escala
train_predictions = scaler.inverse_transform(train_predictions)
test_predictions = scaler.inverse_transform(test_predictions)

# Obtener la última secuencia de prueba para predecir el próximo día
last_sequence = np.expand_dims(test_data[-seq_length:], axis=0)
next_day_prediction = model.predict(last_sequence)
next_day_prediction = scaler.inverse_transform(next_day_prediction)[0]

# Calcular el porcentaje de cambio del precio a lo largo del día
open_price = next_day_prediction[0]
close_price = next_day_prediction[3]
percentage_change = (close_price - open_price) / open_price * 100

# Determinar el valor con mayor precisión comparando las diferencias absolutas
differences = np.abs(next_day_prediction - test_data[-1])
max_index = np.argmax(differences)

# Obtener la fecha del día predicho
fecha_prediccion = df_eth['date'].values[-1] + pd.DateOffset(days=1)

# Imprimir los resultados
print("Fecha del día predicho:", fecha_prediccion.date())
print("Apertura:", next_day_prediction[0])
print("Máximo:", next_day_prediction[1])
print("Mínimo:", next_day_prediction[2])
print("Cierre:", next_day_prediction[3])
print("Porcentaje de cambio del precio a lo largo del día:", percentage_change)
print("Valor con mayor precisión:", ['Máximo', 'Mínimo', 'Cierre'][max_index])
print("Evaluación del modelo LSTM:")
print("Train Loss:", train_loss)
print("Test Loss:", test_loss)
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# DataFrame df_ilv
df_ilv['date'] = pd.to_datetime(df_ilv['date'])
df_ilv.sort_values('date', inplace=True)
data_ilv = df_ilv[['open', 'close', 'high', 'low']].values
scaler_ilv = MinMaxScaler(feature_range=(0, 1))
scaled_data_ilv = scaler_ilv.fit_transform(data_ilv)
train_size_ilv = int(len(scaled_data_ilv) * 0.8)
train_data_ilv = scaled_data_ilv[:train_size_ilv]
test_data_ilv = scaled_data_ilv[train_size_ilv:]

# DataFrame df_rvn
df_rvn['date'] = pd.to_datetime(df_rvn['date'])
df_rvn.sort_values('date', inplace=True)
data_rvn = df_rvn[['open', 'close', 'high', 'low']].values
scaler_rvn = MinMaxScaler(feature_range=(0, 1))
scaled_data_rvn = scaler_rvn.fit_transform(data_rvn)
train_size_rvn = int(len(scaled_data_rvn) * 0.8)
train_data_rvn = scaled_data_rvn[:train_size_rvn]
test_data_rvn = scaled_data_rvn[train_size_rvn:]

# DataFrame df_sc
df_sc['date'] = pd.to_datetime(df_sc['date'])
df_sc.sort_values('date', inplace=True)
data_sc = df_sc[['open', 'close', 'high', 'low']].values
scaler_sc = MinMaxScaler(feature_range=(0, 1))
scaled_data_sc = scaler_sc.fit_transform(data_sc)
train_size_sc = int(len(scaled_data_sc) * 0.8)
train_data_sc = scaled_data_sc[:train_size_sc]
test_data_sc = scaled_data_sc[train_size_sc:]

# DataFrame df_ltc
df_ltc['date'] = pd.to_datetime(df_ltc['date'])
df_ltc.sort_values('date', inplace=True)
data_ltc = df_ltc[['open', 'close', 'high', 'low']].values
scaler_ltc = MinMaxScaler(feature_range=(0, 1))
scaled_data_ltc = scaler_ltc.fit_transform(data_ltc)
train_size_ltc = int(len(scaled_data_ltc) * 0.8)
train_data_ltc = scaled_data_ltc[:train_size_ltc]
test_data_ltc = scaled_data_ltc[train_size_ltc:]

# DataFrame df_ens
df_ens['date'] = pd.to_datetime(df_ens['date'])
df_ens.sort_values('date', inplace=True)
data_ens = df_ens[['open', 'close', 'high', 'low']].values
scaler_ens = MinMaxScaler(feature_range=(0, 1))
scaled_data_ens = scaler_ens.fit_transform(data_ens)
train_size_ens = int(len(scaled_data_ens) * 0.8)
train_data_ens = scaled_data_ens[:train_size_ens]
test_data_ens = scaled_data_ens[train_size_ens:]

# DataFrame df_eth
df_eth['date'] = pd.to_datetime(df_eth['date'])
df_eth.sort_values('date', inplace=True)
data_eth = df_eth[['open', 'close', 'high', 'low']].values
scaler_eth = MinMaxScaler(feature_range=(0, 1))
scaled_data_eth = scaler_eth.fit_transform(data_eth)
train_size_eth = int(len(scaled_data_eth) * 0.8)
train_data_eth = scaled_data_eth[:train_size_eth]
test_data_eth = scaled_data_eth[train_size_eth:]

def create_sequences(data, seq_length):
    X = []
    y = []
    for i in range(len(data)-seq_length):
        X.append(data[i:i+seq_length])
        y.append(data[i+seq_length])
    return np.array(X), np.array(y)

seq_length = 30

# Crear el modelo y hacer predicciones para df_ilv
X_train_ilv, y_train_ilv = create_sequences(train_data_ilv, seq_length)
X_test_ilv, y_test_ilv = create_sequences(test_data_ilv, seq_length)

model_ilv = Sequential()
model_ilv.add(LSTM(64, activation='relu', input_shape=(seq_length, 4)))
model_ilv.add(Dense(4))
model_ilv.compile(optimizer='adam', loss='mse')
model_ilv.fit(X_train_ilv, y_train_ilv, epochs=10, batch_size=32)

last_sequence_ilv = np.expand_dims(test_data_ilv[-seq_length:], axis=0)
next_day_prediction_ilv = model_ilv.predict(last_sequence_ilv)
next_day_prediction_ilv = scaler_ilv.inverse_transform(next_day_prediction_ilv)[0]

# Crear el modelo y hacer predicciones para df_rvn
X_train_rvn, y_train_rvn = create_sequences(train_data_rvn, seq_length)
X_test_rvn, y_test_rvn = create_sequences(test_data_rvn, seq_length)

model_rvn = Sequential()
model_rvn.add(LSTM(64, activation='relu', input_shape=(seq_length, 4)))
model_rvn.add(Dense(4))
model_rvn.compile(optimizer='adam', loss='mse')
model_rvn.fit(X_train_rvn, y_train_rvn, epochs=10, batch_size=32)

last_sequence_rvn = np.expand_dims(test_data_rvn[-seq_length:], axis=0)
next_day_prediction_rvn = model_rvn.predict(last_sequence_rvn)
next_day_prediction_rvn = scaler_rvn.inverse_transform(next_day_prediction_rvn)[0]

# Crear el modelo y hacer predicciones para df_sc
X_train_sc, y_train_sc = create_sequences(train_data_sc, seq_length)
X_test_sc, y_test_sc = create_sequences(test_data_sc, seq_length)

model_sc = Sequential()
model_sc.add(LSTM(64, activation='relu', input_shape=(seq_length, 4)))
model_sc.add(Dense(4))
model_sc.compile(optimizer='adam', loss='mse')
model_sc.fit(X_train_sc, y_train_sc, epochs=10, batch_size=32)

last_sequence_sc = np.expand_dims(test_data_sc[-seq_length:], axis=0)
next_day_prediction_sc = model_sc.predict(last_sequence_sc)
next_day_prediction_sc = scaler_sc.inverse_transform(next_day_prediction_sc)[0]

# Crear el modelo y hacer predicciones para df_ltc
X_train_ltc, y_train_ltc = create_sequences(train_data_ltc, seq_length)
X_test_ltc, y_test_ltc = create_sequences(test_data_ltc, seq_length)

model_ltc = Sequential()
model_ltc.add(LSTM(64, activation='relu', input_shape=(seq_length, 4)))
model_ltc.add(Dense(4))
model_ltc.compile(optimizer='adam', loss='mse')
model_ltc.fit(X_train_ltc, y_train_ltc, epochs=10, batch_size=32)

last_sequence_ltc = np.expand_dims(test_data_ltc[-seq_length:], axis=0)
next_day_prediction_ltc = model_ltc.predict(last_sequence_ltc)
next_day_prediction_ltc = scaler_ltc.inverse_transform(next_day_prediction_ltc)[0]

# Crear el modelo y hacer predicciones para df_ens
X_train_ens, y_train_ens = create_sequences(train_data_ens, seq_length)
X_test_ens, y_test_ens = create_sequences(test_data_ens, seq_length)

model_ens = Sequential()
model_ens.add(LSTM(64, activation='relu', input_shape=(seq_length, 4)))
model_ens.add(Dense(4))
model_ens.compile(optimizer='adam', loss='mse')
model_ens.fit(X_train_ens, y_train_ens, epochs=10, batch_size=32)

last_sequence_ens = np.expand_dims(test_data_ens[-seq_length:], axis=0)
next_day_prediction_ens = model_ens.predict(last_sequence_ens)
next_day_prediction_ens = scaler_ens.inverse_transform(next_day_prediction_ens)[0]

# Crear el modelo y hacer predicciones para df_eth
X_train_eth, y_train_eth = create_sequences(train_data_eth, seq_length)
X_test_eth, y_test_eth = create_sequences(test_data_eth, seq_length)

model_eth = Sequential()
model_eth.add(LSTM(64, activation='relu', input_shape=(seq_length, 4)))
model_eth.add(Dense(4))
model_eth.compile(optimizer='adam', loss='mse')
model_eth.fit(X_train_eth, y_train_eth, epochs=10, batch_size=32)

last_sequence_eth = np.expand_dims(test_data_eth[-seq_length:], axis=0)
next_day_prediction_eth = model_eth.predict(last_sequence_eth)
next_day_prediction_eth = scaler_eth.inverse_transform(next_day_prediction_eth)[0]

# Imprimir los resultados
print("Predicciones para df_ilv:")
print("Fecha de los datos predichos:", df_ilv['date'].iloc[-1])
print("Apertura:", next_day_prediction_ilv[0])
print("Cierre:", next_day_prediction_ilv[1])
print("Máximo:", next_day_prediction_ilv[2])
print("Mínimo:", next_day_prediction_ilv[3])
print()

print("Predicciones para df_rvn:")
print("Fecha de los datos predichos:", df_rvn['date'].iloc[-1])
print("Apertura:", next_day_prediction_rvn[0])
print("Cierre:", next_day_prediction_rvn[1])
print("Máximo:", next_day_prediction_rvn[2])
print("Mínimo:", next_day_prediction_rvn[3])
print()

print("Predicciones para df_sc:")
print("Fecha de los datos predichos:", df_sc['date'].iloc[-1])
print("Apertura:", next_day_prediction_sc[0])
print("Cierre:", next_day_prediction_sc[1])
print("Máximo:", next_day_prediction_sc[2])
print("Mínimo:", next_day_prediction_sc[3])
print()

print("Predicciones para df_ltc:")
print("Fecha de los datos predichos:", df_ltc['date'].iloc[-1])
print("Apertura:", next_day_prediction_ltc[0])
print("Cierre:", next_day_prediction_ltc[1])
print("Máximo:", next_day_prediction_ltc[2])
print("Mínimo:", next_day_prediction_ltc[3])
print()

print("Predicciones para df_ens:")
print("Fecha de los datos predichos:", df_ens['date'].iloc[-1])
print("Apertura:", next_day_prediction_ens[0])
print("Cierre:", next_day_prediction_ens[1])
print("Máximo:", next_day_prediction_ens[2])
print("Mínimo:", next_day_prediction_ens[3])
print()

print("Predicciones para df_eth:")
print("Fecha de los datos predichos:", df_eth['date'].iloc[-1])
print("Apertura:", next_day_prediction_eth[0])
print("Cierre:", next_day_prediction_eth[1])
print("Máximo:", next_day_prediction_eth[2])
print("Mínimo:", next_day_prediction_eth[3])


```
